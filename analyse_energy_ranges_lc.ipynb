```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import astropy.units as u
from datetime import datetime
import pickle, json, sys, os, glob
import pandas as pd
pd.set_option('display.max_columns', None)

from astropy.coordinates import SkyCoord
from scipy.stats import chi2
from scipy.optimize import minimize
from astropy.table import Table
import matplotlib.animation as animation
from matplotlib.animation import FuncAnimation

# location of the scripts
sys.path.insert(0, os.path.join("/fefs/aswg/workspace/juan.jimenez/lst1_systematics/scripts"))
import auxiliar as aux
import geometry as geom

from IPython.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))


# ============================ #
dict_path = "objects/total_dicts"
# ============================ #



#######################################
# intensity integration limits
#######################################

int_lim = [100, 400] #p.h.

#######################################
```

### Reading all the objects and putting all the information in a unique dict

```python
%%time
# reading dicts
dicts = []

# filenames
fnames = np.sort(glob.glob(dict_path + "/*"))
for fname in fnames:

    with open(fname, 'rb') as f:
        dicts.append(pickle.load(f))
```

```python
fluxs, e_fluxs = [], []
e_min, e_max = [], []
chi2val, pvalue = [], []
for d in dicts:
    _f, _e_f,  = [], []
    for run in d["run"].keys():
        _f.append(d["run"][run]["flux"])
        _e_f.append(d["run"][run]["e_flux"])
        
    fluxs.append(_f)
    e_fluxs.append(_e_f)
    e_min.append(d["general"]["e_min_lc"].value)
    e_max.append(d["general"]["e_max_lc"].value)
    chi2val.append(d["general"]["chi2_flux"])
    pvalue.append(d["general"]["pvalue_flux"])
    
```

```python
fig, (ax1) = plt.subplots(figsize=(9, 4))
ax2 = ax1.twinx()
ax3 = ax1.twinx()

p1, = ax1.plot(e_min, chi2val, color="r")
p2, = ax2.plot(e_min, pvalue, color="b")

ax1.set_ylabel("$\chi^2$")
ax2.set_ylabel("p-value")
for ax in [ax1, ax2]:
    ax.set_xlabel("Energy [TeV]")
    ax.set_xscale("log")
    ax.set_xlim(0.008, 2)
             
ax1.yaxis.label.set_color(p1.get_color())
ax2.yaxis.label.set_color(p2.get_color())
ax1.tick_params(axis='y', colors=p1.get_color())
ax2.tick_params(axis='y', colors=p2.get_color())
ax3.set_yticks([])
ax1.set_ylim(140, 240)
ax1.set_title("$\chi^2$ and p-value of fluxes vs a certain energy lower cut")
plt.show()
```

```python
Ibins  = dicts[0]["general"]["intensity_bins"]
IbinsC = dicts[0]["general"]["intensity_bins_centers"]

# select the bins that are inside the integral
flag_bins = []
for i in range(len(Ibins)-1):
    if Ibins[i] > np.log10(int_lim[0]) and Ibins[i+1] < np.log10(int_lim[1]):
        flag_bins.append(True)
    else:
        flag_bins.append(False)
```

### Extracting all the values in arrays

```python
p_rate   = []
zd       = []
e_p_rate = []
for i, run in enumerate(dicts[0]["run"].keys()): 

    d = dicts[0]
    rates = d["run"][run]["intensity_rates"]

    # --- Compute the integral of p rate in the integral range --- #
    integral = np.sum(rates[flag_bins])

    p_rate.append(integral)
    zd.append(d["run"][run]["zd"])
    e_p_rate.append(np.sqrt(d["run"][run]["n_events"]) / d["run"][run]["effective_time"])
```

### Linearly fitting the zd to the proton rate and showing all distributions

```python
# Perform linear fits
fitF_zd_p = np.poly1d(np.polyfit(np.cos(np.deg2rad(zd)), p_rate, 1))


zd_0_p_rate = []
for i, run in enumerate(dicts[0]["run"].keys()):
    
    _zd_0_p_rate = p_rate[i] - fitF_zd_p(np.cos(np.deg2rad(dicts[0]["run"][run]["zd"]))) + fitF_zd_p(1)
    
    zd_0_p_rate.append(_zd_0_p_rate)
```

### Fitting the flux vs the proton rate with a stright line passing trough (0,0)

```python
corr_fluxs = []
corr_chi2, corr_pval, lrt = [], [], []

for i in range(len(fluxs)):
    
    LRT = geom.LRTFreeZero(zd_0_p_rate, fluxs[i], e_fluxs[i])
    
    zd_corr_flux = []
    for j, run in enumerate(dicts[i]["run"].keys()):
        _factor = geom.lineZeroF(zd_0_p_rate[j], LRT[2]) / dicts[i]["run"][run]["flux"]

        _zd_corr_flux = dicts[i]["run"][run]["flux"] - (geom.lineZeroF(zd_0_p_rate[j], LRT[2]) - np.mean(fluxs[i]))

        zd_corr_flux.append(_zd_corr_flux[0])
        
    corr_fluxs.append(zd_corr_flux)
    
    c, _, p = aux.calculate_chi2_pvalue_const(zd_corr_flux, e_fluxs[i])
    corr_chi2.append(c)
    corr_pval.append(p)
    
    lrt.append(np.sqrt(np.abs(c-chi2val[i])))
```

```python
fig, (ax1, axl) = plt.subplots(2, 1, figsize=(13, 5), sharex=True, gridspec_kw={'height_ratios': [2, 1]})

ax2 = ax1.twinx()
ax3 = ax1.twinx()

p1, = ax1.plot(e_min, chi2val, color="r", ls="--", label="Original data")
p2, = ax2.plot(e_min, pvalue, color="b", ls="--")
ax1.plot(e_min, corr_chi2, color="r", ls="-", label="Corrected data")
ax2.plot(e_min, corr_pval, color="b", ls="-")

ax1.set_ylabel("$\chi^2$")
ax2.set_ylabel("p-value")
for ax in [ax1, ax2]:
    ax.set_xscale("log")
    ax.set_xlim(0.008, 2)

    
axl.plot(e_min, lrt, color="g", ls="-",)


axl.set_ylabel(f"LRT significance [{aux.sigma}]")

axl.set_xlabel("Energy [TeV]")
axl.set_xscale("log")
axl.set_xlim(0.008, 2)
axl.set_ylim(2.8, 5.5)


ax1.legend(loc=6)
ax1.yaxis.label.set_color(p1.get_color())
ax2.yaxis.label.set_color(p2.get_color())
ax1.tick_params(axis='y', colors=p1.get_color())
ax2.tick_params(axis='y', colors=p2.get_color())
ax3.set_yticks([])
ax1.set_ylim(120, 240)
plt.subplots_adjust(wspace=0, hspace=0)
ax1.set_title("$\chi^2$ and p-value of fluxes vs a certain energy lower cut")
plt.show()
```
